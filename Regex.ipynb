# **Regular Expressions (regex)** in **Python**
---

## üîπ What is Regex?

**Regular Expression (regex)** is a *pattern* that helps you **search, match, or manipulate text** based on specific rules.

You use it for:

* Finding words or numbers in a string
* Validating emails, phone numbers, passwords
* Replacing or splitting text
* Extracting useful data from messy text

---

## üîπ Importing the Regex Module

In Python, regex is handled using the **`re`** module.

```python
import re
```

---

## üîπ Basic Functions in `re`

| Function       | Description                              | Example                   |
| -------------- | ---------------------------------------- | ------------------------- |
| `re.search()`  | Searches for the **first** match         | `re.search("abc", text)`  |
| `re.findall()` | Finds **all** matches                    | `re.findall("\d+", text)` |
| `re.match()`   | Checks only at the **start** of a string | `re.match("abc", text)`   |
| `re.split()`   | Splits a string by pattern               | `re.split("\s", text)`    |
| `re.sub()`     | Replaces a pattern with something        | `re.sub("\d", "#", text)` |

---

## üîπ Regex Pattern Basics

### 1. **Ordinary characters**

They match themselves.

```python
re.search("cat", "The cat sat")  # ‚úÖ Match found
re.search("dog", "The cat sat")  # ‚ùå No match
```

---

### 2. **Metacharacters (Special symbols)**

These are powerful ‚Äî they let you define flexible patterns.

| Symbol  | Meaning                        | Example     | Matches                       |      |                |
| ------- | ------------------------------ | ----------- | ----------------------------- | ---- | -------------- |
| `.`     | Any character (except newline) | `c.t`       | `cat`, `cut`                  |      |                |
| `^`     | Start of string                | `^Hello`    | `Hello world`                 |      |                |
| `$`     | End of string                  | `world$`    | `Hello world`                 |      |                |
| `*`     | 0 or more repetitions          | `lo*`       | `l`, `lo`, `loo`, `looo`      |      |                |
| `+`     | 1 or more repetitions          | `lo+`       | `lo`, `loo`                   |      |                |
| `?`     | 0 or 1 repetition              | `colou?r`   | `color`, `colour`             |      |                |
| `{n}`   | Exactly *n* times              | `\d{3}`     | `123`                         |      |                |
| `{n,}`  | *n* or more times              | `\d{2,}`    | `12`, `12345`                 |      |                |
| `{n,m}` | Between *n* and *m* times      | `\d{2,4}`   | `12`, `1234`                  |      |                |
| `[]`    | Character class                | `[abc]`     | `a`, `b`, or `c`              |      |                |
| `[^]`   | Negation                       | `[^abc]`    | anything except `a`, `b`, `c` |      |                |
| `       | `                              | OR operator | `cat                          | dog` | `cat` or `dog` |
| `()`    | Grouping                       | `(abc)+`    | `abc`, `abcabc`               |      |                |

---

## üîπ Common **Shorthand Character Classes**

| Symbol | Meaning                             | Example | Matches     |
| ------ | ----------------------------------- | ------- | ----------- |
| `\d`   | Digit                               | `\d+`   | `123`, `45` |
| `\D`   | Non-digit                           | `\D+`   | `abc`, `@!` |
| `\w`   | Word character (letters, digits, _) | `\w+`   | `hello_123` |
| `\W`   | Non-word character                  | `\W+`   | `@!#`       |
| `\s`   | Whitespace (space, tab, newline)    | `\s+`   | space, tab  |
| `\S`   | Non-whitespace                      | `\S+`   | `hello`     |

---

## üîπ Example 1: Extract all numbers

```python
import re
text = "I have 2 apples and 15 bananas"
numbers = re.findall(r"\d+", text)
print(numbers)
```

**Output:**

```
['2', '15']
```

Explanation:
`\d+` ‚Üí one or more digits.

---

## üîπ Example 2: Validate an Email

```python
email = "user.name123@gmail.com"
pattern = r"^[\w\.-]+@[\w\.-]+\.\w+$"

if re.match(pattern, email):
    print("Valid email")
else:
    print("Invalid email")
```

**Explanation:**

* `^` ‚Üí start
* `[\w\.-]+` ‚Üí letters, digits, dot, or dash
* `@` ‚Üí literal @
* `[\w\.-]+` ‚Üí domain name
* `\.\w+` ‚Üí dot followed by letters
* `$` ‚Üí end

---

## üîπ Example 3: Replace all digits with `#`

```python
text = "My number is 12345"
result = re.sub(r"\d", "#", text)
print(result)
```

**Output:**

```
My number is #####
```

---

## üîπ Example 4: Split a sentence into words

```python
text = "Hello, how are you?"
words = re.split(r"\W+", text)
print(words)
```

**Output:**

```
['Hello', 'how', 'are', 'you', '']
```

`\W+` splits by non-word characters (like punctuation and spaces).

---

## üîπ Example 5: Find words starting with capital letters

```python
text = "Alice and Bob are Good Friends"
caps = re.findall(r"\b[A-Z][a-z]*\b", text)
print(caps)
```

**Output:**

```
['Alice', 'Bob', 'Good', 'Friends']
```

**Explanation:**

* `\b` ‚Üí word boundary
* `[A-Z]` ‚Üí first letter capital
* `[a-z]*` ‚Üí followed by zero or more lowercase letters

---

## üîπ Example 6: Extract all hashtags from text

```python
text = "I love #Python and #MachineLearning"
tags = re.findall(r"#\w+", text)
print(tags)
```

**Output:**

```
['#Python', '#MachineLearning']
```

---

## üîπ Example 7: Match phone numbers

```python
text = "Call me at 050-123-4567 or 04-987-6543"
phones = re.findall(r"\d{2,3}-\d{3}-\d{4}", text)
print(phones)
```

**Output:**

```
['050-123-4567', '04-987-6543']
```

## üîπ Flags (Optional Parameters)

Flags modify how regex behaves.

| Flag                      | Name                            | Meaning                         |
| ------------------------- | ------------------------------- | ------------------------------- |
| `re.IGNORECASE` or `re.I` | Ignore case                     | `re.search("cat", "CAT", re.I)` |
| `re.MULTILINE` or `re.M`  | `^` and `$` work for every line |                                 |
| `re.DOTALL` or `re.S`     | `.` matches newline too         |                                 |

**Example:**

```python
re.search("hello", "HELLO", re.I)  # Case insensitive
```

---

## üîπ Example 8: Combining flags

```python
text = """Hello
world"""
match = re.search(r"Hello.*world", text, re.S)
print(bool(match))
```

Without `re.S`, `.` won‚Äôt match newline, so no match.
With `re.S`, it matches across lines.
