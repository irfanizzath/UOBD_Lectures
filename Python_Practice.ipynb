````python
import random

choices = {1: 'Rock', 2: 'Paper', 3: 'Pencil', 4: 'Scissors'}

print('ROCK... PAPER... PENCIL... SCISSOR!!!')
for i, j in choices.items():
    print(i, ':', j)

def play_rpps(choice_num):
    player_choice = choices[choice_num]
    comp_choice = choices[random.randint(1, 4)]

    print(f"\nYou chose: {player_choice}")
    print(f"Computer chose: {comp_choice}\n")

    # Draw case
    if player_choice == comp_choice:
        return 'Draw'

    # Define winning rules (who beats who)
    wins = {
        'Rock': ['Scissors', 'Pencil'],
        'Paper': ['Rock'],
        'Pencil': ['Paper'],
        'Scissors': ['Paper', 'Pencil']
    }

    # Check who wins
    if comp_choice in wins[player_choice]:
        return 'You Win!'
    else:
        return 'You Lose!'

# Run game
result = play_rpps(int(input("Enter your choice (1-4): ")))
print(result)
````

#  **5. Formatting Dates (`strftime`)**

Convert datetime to a string in a specific format.

```python
print(now.strftime("%Y-%m-%d"))        # 2025-12-11
print(now.strftime("%d/%m/%Y %H:%M"))  # 11/12/2025 08:12
print(now.strftime("%A, %B %d, %Y"))   # Thursday, December 11, 2025
```

**Common codes:**

* `%Y`  year, `%y`  2-digit year
* `%m`  month number, `%B`  full month name
* `%d`  day of month
* `%H`  hour (24h), `%I`  hour (12h)
* `%M`  minute, `%S`  second
* `%A`  weekday name, `%a`  short weekday

---

**Read a CSV file into a Python dictionary**.

---

###  **Case 1: CSV with header row (like a table)**

If your CSV looks like this 

```
Name,Age,Country
Irfan,25,UAE
Sara,28,India
Ali,30,Saudi Arabia
```

You can use `csv.DictReader()`:

```python
import csv

with open("people.csv", "r") as file:
    reader = csv.DictReader(file)
    for row in reader:
        print(row)
```

 **Output:**

```python
{'Name': 'Irfan', 'Age': '25', 'Country': 'UAE'}
{'Name': 'Sara', 'Age': '28', 'Country': 'India'}
{'Name': 'Ali', 'Age': '30', 'Country': 'Saudi Arabia'}
```

 Each row becomes a **dictionary**, where keys are the **column headers**.

---

###  **Case 2: Simple keyvalue CSV**

If your CSV looks like this 

```
Key,Value
Name,Irfan
Age,25
Country,UAE
```

You can read it into a single dictionary:

```python
import csv

data = {}

with open("output.csv", "r") as file:
    reader = csv.reader(file)
    next(reader)  # skip header row
    for key, value in reader:
        data[key] = value

print(data)
```

 **Output:**

```python
{'Name': 'Irfan', 'Age': '25', 'Country': 'UAE'}
```

---

###  Quick Tip:

If you want all values in correct types (like numbers), you can cast them:

```python
data[key] = int(value) if value.isdigit() else value
```


### **1. Class**

A blueprint for creating objects. It defines attributes (variables) and methods (functions) that the objects will have.

**Example:**

```python
class Car:
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model
```

---

### **2. Object**

An instance of a class. Objects hold specific data and can perform actions defined in the class.

**Example:**

```python
my_car = Car("Toyota", "Corolla")
```

---

### **3. Attribute**

A variable associated with a class or object. Attributes store data about the object.

**Example:**

```python
my_car.brand  # "Toyota"
```

---

### **4. Method**

A function defined inside a class that can operate on object attributes.

**Example:**

```python
class Car:
    def start_engine(self):
        print("Engine started")

my_car.start_engine()  # Engine started
```

---

### **5. Constructor (`__init__`)**

A special method that is called when an object is created. Used to initialize attributes.

**Example:**

```python
def __init__(self, brand, model):
    self.brand = brand
    self.model = model
```

---

### **6. Inheritance**

The mechanism by which one class (child/subclass) can acquire the properties and methods of another class (parent/superclass).

**Example:**

```python
class Vehicle:
    def move(self):
        print("Moving")

class Car(Vehicle):
    pass

my_car = Car()
my_car.move()  # Moving
```

---

### **7. Single Inheritance**

A child class inherits from **one parent class**.

---

### **8. Multiple Inheritance**

A child class inherits from **more than one parent class**.

**Example:**

```python
class A:
    pass
class B:
    pass
class C(A, B):
    pass
```

---

### **9. Multilevel Inheritance**

A class is derived from a child class, which is derived from another parent class.

**Example:**

```python
class A:
    pass
class B(A):
    pass
class C(B):
    pass
```

---

### **10. Hierarchical Inheritance**

Multiple child classes inherit from **the same parent class**.

---

### **11. Encapsulation**

The concept of **hiding internal details** of a class and exposing only whats necessary. Usually done using **private (`__var`) or protected (`_var`) attributes**.

**Example:**

```python
class Person:
    def __init__(self, name):
        self.__name = name  # private attribute
```

---

### **12. Abstraction**

The concept of **hiding complex implementation details** and showing only essential features.

**Example:**
Using abstract classes or interfaces:

```python
from abc import ABC, abstractmethod

class Vehicle(ABC):
    @abstractmethod
    def move(self):
        pass
```

---

### **13. Polymorphism**

The ability of objects of **different classes** to respond to the same method call in different ways.

**Example:**

```python
class Cat:
    def sound(self):
        print("Meow")

class Dog:
    def sound(self):
        print("Bark")

def make_sound(animal):
    animal.sound()

make_sound(Cat())  # Meow
make_sound(Dog())  # Bark
```

---

### **14. Method Overriding**

A child class provides a **different implementation** of a method already defined in the parent class.

**Example:**

```python
class Vehicle:
    def move(self):
        print("Vehicle moving")

class Car(Vehicle):
    def move(self):
        print("Car driving")

Car().move()  # Car driving
```

---

### **15. Method Overloading (Python doesnt support directly)**

Having multiple methods with the same name but **different parameters**.
In Python, can be mimicked using default arguments or `*args`.

**Example:**

```python
class Math:
    def add(self, a, b=0):
        return a + b
```

---

### **16. Property Decorator**

A way to **use methods like attributes**. Provides controlled access to attributes.

**Example:**

```python
class Circle:
    def __init__(self, radius):
        self._radius = radius

    @property
    def radius(self):
        return self._radius

c = Circle(5)
print(c.radius)  # 5, accessed like an attribute
```

---

### **17. Getter and Setter**

Methods to **get or set the value** of private attributes. Often used with `@property`.

---

### **18. Composition**

Building classes by **including objects of other classes** rather than inheriting.

**Example:**

```python
class Engine:
    def start(self):
        print("Engine started")

class Car:
    def __init__(self):
        self.engine = Engine()

Car().engine.start()
```

---

### **19. Association**

A relationship between classes where one class **uses or knows about another**.

---

### **20. Aggregation**

A special form of association where one class **owns another**, but the owned object can exist independently.

---

### **21. Object Identity**

Every object has a unique **ID** in memory. Even two objects with the same data are different.

---

### **22. Self**

A reference to the **current instance** of the class. Always used as the first parameter in class methods.

---

### **23. Super**

Used to **call a method from the parent class** inside a child class.

**Example:**

```python
class Vehicle:
    def move(self):
        print("Vehicle moving")

class Car(Vehicle):
    def move(self):
        super().move()
        print("Car driving")

Car().move()
```

````python
class Age:
  def __init__(self, age):
    self.__age = age

  @property
  def age(self):
    return self.__age

  @age.setter
  def age(self, age):
    if age < 0:
      raise ValueError("Age cannot be negative")
    self.__age = age

  @age.deleter
  def age(self):
    del self.__age
````

````python
from datetime import datetime
import math


class Vehicle:
    def __init__(self, vehicle_id, location, rate):
        self.vehicle_id = vehicle_id
        self.location = location
        self.available = True
        self.rate = rate

    def rent_vehicle(self):
        self.available = False

    def return_vehicle(self, return_station):
        self.available = True
        self.location = return_station

    def refuel(self, amount):
        pass


class Motorbike(Vehicle):
    def __init__(self, vehicle_id, location):
        super().__init__(vehicle_id, location, 1.5)
        self.capacity = 80
        self.fuel_level = 0

    def refuel(self, amount):
        self.fuel_level = min(self.fuel_level + amount, self.capacity)
        print(f'Your motorbike has been refueled to {self.fuel_level}L')


class ElectricScooter(Vehicle):
    def __init__(self, vehicle_id, location):
        super().__init__(vehicle_id, location, 1.1)
        self.battery_level = 10

    def refuel(self, amount):
        if self.check_battery_low():
            self.battery_level = min(self.battery_level + amount, 100)
            print(f'Your scooter has been recharged to {self.battery_level}%')

    def check_battery_low(self):
        if self.battery_level < 20:
            print('ALERT: Time to recharge!')
            return True
        else:
            print(f"Battery level is at {self.battery_level}%, so your scooter doesn't need charging")
            return False


class VehicleRental:
    motorbikes = [
        Motorbike('MB1', 'Selly Oak'),
        Motorbike('MB2', 'Bournville'),
        Motorbike('MB3', 'Harborne')
    ]
    scooters = [
        ElectricScooter('ES1', 'Selly Oak'),
        ElectricScooter('ES2', 'Bournville'),
        ElectricScooter('ES3', 'Harborne')
    ]

    def __init__(self):
        self.vehicle = None
        self.depart_station = None
        self.rent_time = None
        self.return_station = None
        self.return_time = None

    def rent_vehicle(self, vehicle):
        self.vehicle = vehicle
        self.vehicle.rent_vehicle()
        self.depart_station = vehicle.location
        self.rent_time = datetime.now()
        print(f'You have successfully rented vehicle {self.vehicle.vehicle_id} from {self.depart_station}')

    def return_vehicle(self, return_station):
        fare = 0
        self.return_station = return_station
        self.return_time = datetime.now()
        self.vehicle.return_vehicle(return_station)
        fare = self.calculate_fare()
        print(f'You have successfully returned vehicle {self.vehicle.vehicle_id} to {self.return_station}')
        print(f'The fare for this rental is {fare:.2f}')

    def calculate_fare(self):
        fare = 0
        total_minutes = math.ceil((self.return_time - self.rent_time).total_seconds() / 60)
        fare = total_minutes * self.vehicle.rate
        return fare


class User:
    def __init__(self, user_id, name):
        self.user_id = user_id
        self.name = name
        self.current_rental = None
        self.rental_history = []

    def rent_vehicle(self, vehicle):
        self.current_rental = VehicleRental()
        self.current_rental.rent_vehicle(vehicle)
        self.rental_history.append(self.current_rental)

    def return_vehicle(self, return_station):
        self.current_rental.return_vehicle(return_station)
        self.current_rental = None

    def display_rental_history(self):
        print(f'Rental history for {self.name} with ID: {self.user_id}')
        for rental in self.rental_history:
            print(f'Vehicle {rental.vehicle.vehicle_id}')
            print(f'  Rented from: {rental.depart_station} at {rental.rent_time}')
            if rental.return_station:
                print(f'  Returned to: {rental.return_station} at {rental.return_time}')
            else:
                print(f'  Status: Currently rented')
            print('-' * 55)


def display_available_vehicles():
    for motorbike in VehicleRental.motorbikes:
        if motorbike.available:
            print(f'Motorbike {motorbike.vehicle_id} is available from {motorbike.location}')

    for scooter in VehicleRental.scooters:
        if scooter.available:
            print(f'Scooter {scooter.vehicle_id} is available from {scooter.location}')


def main():
    users = {
        'U1': User('U1', 'Evie'),
        'U2': User('U2', 'Nathan')
    }
    current_user = None
    while True:
        if not current_user:
            user_id = input('Please enter your user ID: ')
            if user_id not in users:
                print(f'No user exists with ID {user_id}')
                continue
            current_user = users[user_id]
            print(f'Welcome {current_user.name}')

        try:
            choice = int(input('Book a Motorbike or an Electric Scooter\n'
                               '1. Display available vehicles\n'
                               '2. Rent a vehicle\n'
                               '3. Return a vehicle\n'
                               '4. Refuel / Recharge a vehicle\n'
                               '5. Display user history\n'
                               '6. Logout\n'
                               '7. Exit\n'))
        except ValueError:
            print('You must enter a number from 1-7')
            continue

        match choice:
            case 1:
                display_available_vehicles()
            case 2:
                if not current_user.current_rental:
                    borrow_station = input('Please enter your current station: ')
                    vehicle_type = input('Would you like to rent a (b)ike or a (s)cooter: ')
                    if vehicle_type == 'b':
                        vehicles = VehicleRental.motorbikes
                    elif vehicle_type == 's':
                        vehicles = VehicleRental.scooters
                    else:
                        print("Invalid option, you must pick 'b'(bike) or 's'(scooter)")
                        continue

                    available_vehicles = [vehicle for vehicle in vehicles
                                          if vehicle.available and vehicle.location == borrow_station]

                    if not available_vehicles:
                        print(f"No {'motorbikes' if vehicle_type == 'b' else 'scooters'} are available to rent "
                              f"at {borrow_station}")
                        continue
                    current_user.rent_vehicle(available_vehicles[0])
                else:
                    print("You've already got a vehicle on loan")

            case 3:
                if current_user.current_rental:
                    return_station = input('Please enter the returning station: ')
                    current_user.return_vehicle(return_station)
                else:
                    print("You've not got a vehicle on loan, so can't return anything")
            case 4:
                if current_user.current_rental:
                    current_user.current_rental.vehicle.refuel(50)
                else:
                    print("You've not got a vehicle on loan, so can't refuel/recharge anything")
            case 5:
                current_user.display_rental_history()
            case 6:
                current_user = None
            case 7:
                break
            case _:
                print('Invalid choice entered. You must enter a number from 1-7')


if __name__ == '__main__':
    main()
````
